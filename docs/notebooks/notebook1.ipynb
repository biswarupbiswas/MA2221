## Piecewise Functions

Piecewise-defined functions are written using the `cases` environment.

For example,

$$
f(x) =
\begin{cases}
x^2, & x \ge 0, \\
-x,  & x < 0.
\end{cases}
$$

This is written as

```latex
f(x) =
\begin{cases}
x^2, & x \ge 0, \\
-x,  & x < 0.
\end{cases}
